```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
response_next = requests.get('https://www.epicurious.com/recipes/food/views/butterflied-chicken-with-herbs-and-cracked-olives')
soup_next = BeautifulSoup(response_next.text)
recipe_next_details = {}
recipe_next_details = {'title': soup_next.find('h1',itemprop="name").get_text().strip()}
recipe_next_details["author_name"] = soup_next.find(class_ = 'contributor').get_text().strip()
recipe_next_details["available_since"] = soup_next.find(class_ = 'pub-date').get_text().strip()
recipe_next_details["rating"] = soup_next.find(class_ = 'rating').get_text().strip()
recipe_next_details["description"] = soup_next.find(class_ = 'dek').get_text().strip()
ingredients_html_list = soup_next.find_all(class_ = 'ingredient')
ingr_list = []
for ingr in ingredients_html_list:
    ingr_list.append(ingr.get_text().strip())
ingr_list
recipe_next_details["ingredient_list"] = ingr_list
prep_html = soup_next.find_all(class_ = 'preparation-step')
prep_list = []
for prep in prep_html:
    prep_list.append(prep.get_text().strip())
recipe_next_details["preparation_steps"] = prep_list
review_html = soup_next.find_all(class_ = 'review-text')
review_list = []
for review in review_html:
    review_list.append(review.get_text().strip())
recipe_next_details["reviews"] = review_list
pd.DataFrame.from_dict(recipe_next_details, orient='index').reset_index()
```
